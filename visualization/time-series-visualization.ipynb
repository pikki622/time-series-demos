# Time series visualization

Based on https://www.digitalocean.com/community/tutorials/a-guide-to-time-series-visualization-with-python-3

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

### Loading data

```python
data = sm.datasets.co2.load_pandas()
co2 = data.data
```

```python
co2.head(5)
```

### Indexing

```python
co2.index
```

```python
y = co2['co2'].resample('MS').mean()
```

```python
y.head(5)
```

```python
y['1990':]
```

```python
y['1995-10-01':'1996-10-01']
```

### Missing values

```python
y.isnull().sum()
```

```python
y = y.fillna(y.bfill())
```

```python
y.isnull().sum()
```

### Visualize

```python
y.plot(figsize=(15,6))
plt.show()
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 11, 9

decomposition = sm.tsa.seasonal_decompose(y, model='additive')
fig = decomposition.plot()
plt.show()
```
